```python
ax=df.query('Location != "Other"')\
    .groupby('Location')['Speed_mph']\
        .agg(['mean','count'])\
            .query('count >= 8')\
                .sort_values('mean')['mean']\
                .plot(kind='barh',figsize=(12,5),title='Velocidad promedio por localización')

ax.set_xlabel('Average Coaster Speed')
plt.show()
```
